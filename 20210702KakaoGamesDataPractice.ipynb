카카오게임즈의 '2021년' 정보만 빼서 새로운 dataframe으로 보여주세요.

일단 함수를 써서 정보를 가져올 수 있습니다. 함수를 써서 테이블의 '일자' 열에서 .split()를 써서 '/' 기준으로 년도를 '2021'으로 구분하면 됩니다.

```python
def date(string):
    if '2021' == string.split('/')[0]:
        return True
    else :
        return False
```

새로 만든 함수를 응용하면 됩니다.

```python
df[df['일자'].apply(date)]
```

고가에서 저가를 뺀 값을 dataframe에 새로운 열을 추가해주세요.

```python
df['고가-저가'] = df['고가'] - df['저가']
df
```

추가한 열을 다시 뺄려면 .drop() method를 쓰고 axis는 0이 행, 1이 열이고, 'inplace=True'를 해서 변경된 사항을 저장하면 됩니다.

```python
df.drop("고가-저가",axis=1,inplace=True)
df
```
