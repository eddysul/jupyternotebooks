Use the .dropna() method to drop null values. Set 'how' to 'all' to get rid of rows that only have null values for all of its values. Set 'inplace' as 'True' to keep the changes.

Find how many occupation titles have the title "managers" in them.

First, create function that checks whether "managers" is in the string.

```python
def managers(string):
    if 'managers' in string.lower().split():
        return True
    else:
        return False
```

Apply the function on the occupation title.

```python
df[0]['Occupation title (click on the occupation title to view its profile)'].apply(managers)
```

Use sum to find the total amount of 'true' boolean values

```python
sum(df[0]['Occupation title (click on the occupation title to view its profile)'].apply(managers))
```
