Import Pandas

```python
import pandas as pd
```

Read an external csv file into python and save as variable.

Here is a CSV file containing information about COVID 19 Geographical Distribution in 2020.

```python
df = pd.read_csv('COVID geodist')
df
```

Find column entries and information about dataframe.

```python
df.info()
```

Find how many unique entries there are of countries and territories.

```python
df['countriesAndTerritories'].nunique()
```

Index to find first 10 values of United States of America. *Note that between space there is underscore so you would have to type 'United_States_of_America' in order to index for USA. 

Set a condition to where countries are the USA and find all the instances where the condition is met.

```python
df['countriesAndTerritories'] == 'United_States_of_America'
```

Shows portion of dataframe where the boolean value is true

```python
df[df['countriesAndTerritories'] == 'United_States_of_America']
```

Use .head() method to find the first couple values and 10 in the parentheses to get the first 10 values.

```python
df[df['countriesAndTerritories'] == 'United_States_of_America'].head(10)
```

Find the total number of deaths in South Korea.

First get table with all values of South Korea.

```python
df[df['countriesAndTerritories'] == 'South_Korea']
```

Then get the sum by using .sum() method and gives information on all sums of the data

```python
df[df['countriesAndTerritories'] == 'South_Korea'].sum()
```

To get the numerical value index for deaths in the object select the 'deaths' portion of the table.

```python
df[df['countriesAndTerritories'] == 'South_Korea'].sum()['deaths']
```

Find number of countries where total deaths in 2020 have exceeded 1000 per country.

Since there are multiple entries for each country, using the groupby() method will be beneficial to group all entries into one.

First use the .groupby() method to group by countries and territories and it will give you an object.

```python
df.groupby('countriesAndTerritories')
```

Then use the .sum() method and index for 'deaths' to get the total number of deaths for each country.

```python
df.groupby('countriesAndTerritories').sum()['deaths']
```

Find number of areas where deaths have exceeded 1000 per country by placing a condition and then the sum to find the total number of countries that fulfill the condition

```python
sum(df.groupby('countriesAndTerritories').sum()['deaths'] > 1000)
```

Make pivot table that has continent and countries as index and cases and deaths as values.

Use the .pivot_table and set values as 'cases' and 'deaths' and index as 'continentExp' and 'countriesAndTerritories'

```python
df.pivot_table(values=['cases','deaths'],index=['continentExp','countriesAndTerritories'])
```
